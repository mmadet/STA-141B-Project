```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from datetime import datetime
import plotly
ufo = pd.read_csv("ufos_complete.csv")
#ufo
date_seen = ufo["date_seen"]
date_seen = pd.to_datetime(date_seen)
ufo.head()
type(date_seen)
```

```python
import matplotlib.pyplot as plt
umonth = date_seen.dt.month
hour = date_seen.dt.hour
top_month= umonth.value_counts(sort=False, ascending=False).to_frame()#which month has most ufo sightings
x = date_seen[date_seen.dt.year<=2015]
x = x.dt.date
y = np.random.randint(3000, size=x.shape)
 #test[1,:]
fig = plt.figure(figsize=(18,10))
plt.title('UFO Sightings by Month')
plt.ylabel('Number of UFO Sightings')
plt.xlabel('Month')
plt.plot(top_month)
plt.show()
```

```python
days = x.value_counts(sort = False, ascending = False)
fig = plt.figure(figsize=(18,10))
matplotlib.rcParams.update({'font.size': 22})
plt.title("UFO Sightings through the Years")
plt.ylabel("Number of UFO Sightings")
plt.xlabel("Days")
days.plot()
plt.show()
```

```python
date_posted = ufo["date_posted"]
date_posted = pd.to_datetime(date_posted)
date_posted.head() 
```

```python
#remove records that are invalid because years are 2065,2066, etc..
a= date_seen[date_seen.dt.year<=2015]
b = date_posted[date_posted.dt.year<=2015]
r = np.mean(abs(b-a))
print r #averagenumber of days between date seen and date posted
#a = pd.DataFrame(a)
#b = pd.DataFrame(b)
x2 = np.asarray(sorted(b))
x = a.dt.date
y = range(len(x))
y2 = range(len(x2))
fig = plt.figure(figsize=(18,10))
plt.title("UFO Date Seen and Date Posted", fontsize = 20)
plt.xlabel("Year", fontsize = 18)
plt.ylabel("Number of Sightings", fontsize = 18)
plt.plot(x,y, label = "Seen")
plt.plot(x2,y2, label = "Posted")
plt.legend( loc='upper left', numpoints = 1 )
fig = matplotlib.pyplot.gcf()
matplotlib.rcParams.update({'font.size': 22})
plt.show()
```
