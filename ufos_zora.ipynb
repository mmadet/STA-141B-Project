```python
#read in data file disasters.csv
import pandas as pd
import plotly
from plotly.graph_objs import *
plotly.offline.init_notebook_mode(connected=True)
```

```python
ufos = pd.read_csv('ufos_complete.csv',header =0,dtype=object)
            
#
ufos=ufos.iloc[0:67346]  #remove records of 2014
print ufos.info()
ufos.tail()
```

```python
by_state=dict.fromkeys(list(set(ufos['state_abbr'])))
for key in by_state:
    by_state[key]=len(ufos[ufos['state_abbr']==key]) 
sum(by_state[value] for value in by_state)
```

```python
by_state_df=pd.DataFrame(by_state.items(),columns=['state_abbr','count'])
by_state_df.head()
len(by_state_df)
```

```python
states=dict(set(zip(ufos['state'],ufos['state_abbr'])))
```

```python
states_df=pd.DataFrame(states.items(),columns=['state','state_abbr'])
states_df.head()
len(states_df)
```

```python
counts_state=pd.merge(by_state_df,states_df,on='state_abbr')
counts_state.head()
```

```python
type(by_state_df['state_abbr'][0])
```

```python
scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]

counts_state['text'] =counts_state['state'] 
```

```python
data = [ dict(
        type='choropleth',
       # colorscale = scl,
        autocolorscale = True,
        locations = counts_state['state_abbr'],
        z = counts_state['count'].astype(float),
        locationmode = 'USA-states',
        text = counts_state['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "times")
        ) ]

layout = dict(
        title = '1965-2013 US UFO Sightings by State<br>',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
plotly.offline.iplot( fig, filename='ufos_by_state' )
```

```python
by_latlon=dict.fromkeys(list(set(zip(ufos['longitude'],ufos['latitude']))))
```

```python

for key in by_latlon:
    by_latlon[key]=len(ufos[(ufos['longitude']==key[0])&(ufos['latitude']==key[1])]) 
#print by_latlon
```

```python
#sum(by_latlon[value] for value in by_latlon)
```

```python
by_latlon_df=pd.DataFrame(by_latlon.items(),columns=['latlon','count'])
by_latlon_df['lon'],by_latlon_df['lat'] = zip(*by_latlon_df['latlon'])
```

```python
print by_latlon_df.head()
by_latlon_df[by_latlon_df['count']==max(by_latlon_df['count'])]
```

```python
by_latlon_df.to_csv('ufos_by_latlon.csv')
```
