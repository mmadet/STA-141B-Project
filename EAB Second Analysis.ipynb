```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from datetime import datetime
from pandas import Series, DataFrame, Panel
```

```python
ufo = pd.read_csv("ufos_complete.csv")
disaster = pd.read_csv("disasters.csv")
```

```python
# WHAT TOP COUNTY HAS BEEN AFFECTED MOST BY DISASTERS

# Based on US ufo sighting heat map, check CA, NY, WA, AZ, TX, FL
# Top county is Los Angeles County with 40 hits

ca_dist = disaster[disaster['state'] == 'California']
ca_dist['county'].value_counts()[0:5]
```

```python
# Reported UFO shapes

ca = ufo[ufo['state'] == 'California']
la = ca[ca['county'] == 'Los Angeles County']
shapecount = la['shape'].value_counts()
shapecount.plot.barh()
plt.show()
```

```python
# WHAT COUNTY HAS THE MOST REPORTED UFO SIGHTINGS
# Los Angeles has the most UFO sightings
# What disasters occur there?

ca_dis = disaster[disaster['state'] == 'California']
la_dis = ca_dis[ca_dis['county'] == 'Los Angeles County']
dis_count = la_dis['disaster_type'].value_counts()
dis_count.plot.barh()
plt.show()
```
